## 2.1（2分）对poem.xlsx中的诗歌进行向前最大匹配分词，词表使用上一问中的word_list，最大长度使用上一问中的max_length。

```python
# # 实现前向最大匹配分词算法
# # 要求去除#$ （空格）这三种符号，去除括号内类似（一作 ***）的内容，不必需考虑括号重叠（（））括号缺少）的情况
# 
notation = ['#', '$', ' ']
def cut(string, word_list, max_length):
    # string: 待分词的串
    # word_list: 词表
    # max_length: 词表中词语的最大长度
    # TODO
    cut_list = []
    while string != "":
        for length in range(max_length, 0, -1):
            if string[:length] in word_list or length == 1:
                if string[:length] in notation:
                    string = string[length:]
                elif string[:length] == '(':
                    dis = 0
                    while length + dis + 1 <= len(string) and string[length + dis:length + dis + 1] != ')':
                        dis += 1
                    if length + dis + 1 <= len(string):
                        string = string[length + dis + 1:]
                    else:
                        return ["match error"]
                else:
                    cut_list.append(string[:length])
                    string = string[length:]
                break
    print(" ".join(cut_list))
    return cut_list


table['words'] = table['content'].apply(lambda x: ' '.join(cut(x, word_list, max_length)))
print(table)
```

```python
# IDF 2分
# TODO
import math
poem_num = len(set(new_data['Poem_id'].values))
res2 = new_data.groupby(['word']).agg({'Poem_id':lambda x:len(set(x.values))}).reset_index().rename(columns={'Poem_id':'Poem_num'})
res2['IDF']=res2['Poem_num'].apply(lambda x:math.log(poem_num/x))
res2 = res2.drop(columns='Poem_num')
new_data=new_data.merge(res2,on=['word'],how='left')

new_data['TF-IDF'] = new_data['TF'] * new_data['IDF']
print(new_data)
```

```python
# TODO
def cosine_similarity(a, b):
    # TODO
    a1 = np.sqrt(np.sum(a * a))
    b1 = np.sqrt(np.sum(b * b))

    cos = np.matmul(a.T, b) / (a1 * b1)
    return cos

cosine_similarity(new_data.loc[0]['Poem_vec'], new_data.loc[2500]['Poem_vec'])

def similarity(a):
    sim_matrix = []
    for i in a:
        for j in a:
            sim_matrix.append(cosine_similarity(i, j))
    return np.array(sim_matrix).reshape(300, 300)

pd.DataFrame(similarity(new_data.groupby('Poem_id').head(1)['Poem_vec']), index=new_data['Poem_id'].unique(),
             columns=new_data['Poem_id'].unique())
```

## 3.2（1分）（选做2）实现基于词典的分词方法SPFA。

```python
import queue


class Vertex:
    def __init__(self, key):
        self.id = key
        self.connectedTo = {}

    def add_neighbor(self, nbr, weight=0):
        self.connectedTo[nbr] = weight

    def __str__(self):
        return str(self.id) + ' connectedTo ' + str([x for x in self.connectedTo])

    def get_connections(self):
        return self.connectedTo.keys()

    def get_id(self):
        return self.id

    def get_weight(self, nbr):
        return self.connectedTo[nbr]


class Graph:
    def __init__(self, num=None):
        self.vertList = {}
        self.numVertices = 0
        if num is not None:
            [self.add_vertex(i) for i in range(num)]

    def add_vertex(self, key):
        self.numVertices += 1
        v = Vertex(key)
        self.vertList[key] = v
        return v

    def get_vertex(self, n):
        if n in self.vertList:
            return self.vertList[n]
        else:
            return None

    def __contains__(self, item):
        return item in self.vertList

    def add_edge(self, u, v, w=float('inf')):
        # u到v的weight
        if u not in self.vertList:
            self.add_vertex(u)
        if v not in self.vertList:
            self.add_vertex(v)
        self.vertList[u].add_neighbor(self.vertList[v].get_id(), w)

    def get_vertices(self):
        return self.vertList.keys()

    def __iter__(self):
        return iter(self.vertList.values())


def tokenize(sentence, path):
    token = []
    idx = len(path) - 1
    while idx:
        token.append(sentence[path[idx]:idx])
        idx = path[idx]
    token.reverse()
    return token


def spfa(graph):
    if graph is None:
        return None
    dis = [float('inf') if i != 0 else 0 for i in range(graph.numVertices)]
    path = [0 for i in range(graph.numVertices)]
    book = [False for i in range(graph.numVertices)]
    que = queue.Queue()
    que.put(graph.get_vertex(0))
    while not que.empty():
        u = que.get()
        u_id = u.get_id()
        con = [nei for nei in u.get_connections()]
        for c in con:
            if dis[c] > dis[u_id] + u.get_weight(c):
                dis[c] = dis[u_id] + u.get_weight(c)
                path[c] = u_id
                if not book[c]:
                    que.put(graph.get_vertex(c))
                    book[c] = True

        book[u_id] = False

    return path

# 按照PPT的例子，对下面这句话进行分词
# 老师说明天下午休息
sentence = "老师说明天下午休息"
graph = Graph(9)
# 可以理解成左闭右开
graph.add_edge(0, 1, 1)
graph.add_edge(1, 2, 1)
graph.add_edge(2, 3, 1)
graph.add_edge(3, 4, 1)
graph.add_edge(4, 5, 1)
graph.add_edge(5, 6, 1)
graph.add_edge(6, 7, 1)
graph.add_edge(7, 8, 1)
graph.add_edge(8, 9, 1)
graph.add_edge(0, 2, 1)
graph.add_edge(2, 4, 1)
graph.add_edge(4, 6, 1)
graph.add_edge(3, 5, 1)
graph.add_edge(5, 7, 1)
graph.add_edge(7, 9, 1)
path = spfa(graph)
print(path)
print(tokenize(sentence, path))
```
